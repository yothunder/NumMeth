# 8 - COMPLEXITY

## Summary
1. The complexity of an algorithm is the relationship between the size of the input problem and the time it takes for the algorithm to terminate.
2. Big-O notation is a standard method of classifying algorithmic complexity in a way that is computer- and operating-system-independent.
3. Algorithms with log-complexity perform faster than algorithms with polynomial complexity, which are faster than algorithms with exponential complexity.
4. The Python profiler is a useful tool for determining where your code is running slowly so that you can improve its performance.
