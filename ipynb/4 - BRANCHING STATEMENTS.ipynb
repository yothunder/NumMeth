# 4 - BRANCHING STATEMENTS

## Summary
1. Branching (if-else) statements allow functions to take different actions under different circumstances.
2. Ternary operators allow single line branching statements.

```python
def my_tip_calc(bill,party):
    if party <= 6:
        tips = bill*(15/100)
    elif party <= 8 and party > 6:
        tips = bill*(18/100)
    elif party <= 11 and party > 8:
        tips = bill*(20/100)
    elif party >11:
        tips = bill*(25/100)
    return tips

bill  = 109.29
party = [3, 7, 9, 12]
t     = []
for i in party:
    tt = my_tip_calc(bill,i)
    t.append(tt)
print(t)
```

```python
import numpy as np

def my_mult_operation(a,b,operation):
    if operation == "plus":
        out = a+b
    elif operation == "minus":
        out = a-b
    elif operation == "mult":
        out = a*b
    elif operation == "div":
        out = a/b
    elif operation == "pow":
        out = a**b
    print(out)
    return out

opr = str(input("Pilih satu operasi berikut: [plus, minus, mult, div, pow]"))
x = np.array([1,2,3,4])
y = np.array([2,3,4,5])
print(f'\nUntuk operasi {opr}, hasil perhitungan adalah sebagai berikut')
my_mult_operation(x,y,opr)
```
